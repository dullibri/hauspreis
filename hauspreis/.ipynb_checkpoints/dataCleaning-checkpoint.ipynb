```python
import pandas as pd
import re
import itertools
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
df = pd.read_json("expose.json")
```

```python
#df.info()
```

```python
def cleaner(x):
    sub_list = [('\r\n|m²|\s|€|\.',""),(",",".")]
    for a,b in sub_list:
        x = re.sub(a,b,x)

    return x
```

```python
df.columns
```

```python
simpleList = ['anzahl_raeume','grundstuecksflaeche','kaufpreis','wohnflaeche']
for a in simpleList:
    df[a] = df[a].apply(cleaner)
df.head()
```

```python
df['plz']=df['ort'].apply(lambda x: re.findall('\d{5}',x.split(',')[0])[0])
```

```python
re.sub('^, ','',df.merkmale[1])
```

```python
df['merkmale'] = df['merkmale'].apply(lambda x: re.sub('^, ','',x) if x is not None else x)
```

```python
df['merkmale'] = df.merkmale.apply(lambda x: x.split(',') if x is not None else x)
```

```python
df['merkmale'] = df.merkmale.apply(lambda x: [re.sub('^\s','',a) for a in x] if x is not None else x)
```

```python
df.merkmale[0]
```

```python
merkmalListe = list(np.unique([*itertools.chain.from_iterable(df.merkmale.dropna())]))
```

```python
merkmalListe = filter(lambda x:len(x)>0,merkmalListe)
merkmalListe = [a for a in merkmalListe]
merkmalListe
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
# Manche Exposes haben keine Merkmalbeschreibung. Da die unten stehende Dummy-Generierung
# eine list als Input braucht, macht man hier aus None kurzerhand eine leere list 
misMerkmRow = df.merkmale[df.merkmale.apply(lambda x: x==None)]
df.merkmale[misMerkmRow.index]=''
```

```python
mlb = MultiLabelBinarizer()

df = pd.merge(df
               ,pd.DataFrame(mlb.fit_transform(df.merkmale)
                             ,columns=mlb.classes_
                             ,index=df.index)
              ,left_index=True, right_index=True
             )
              
```

```python
df.rename(columns={"":"keine merkmale"},inplace=True)
```

```python
df.columns
```

```python
df = df.drop(['merkmale'],axis=1)
```

```python
list(df.stadteilbewertung.unique())
```

```python
def staBew(x):
    if x==None:
        return None
    else:
        return re.findall('\d',x)[0]
```

```python
df['stadteilbewertung'] = df['stadteilbewertung'].apply(staBew)
```

```python
df['energie']=df[['energielabel','energiemerkmale']].apply(lambda row:{a:b for a,b in 
                                                         zip(row['energielabel']
                                                             ,row['energiemerkmale'])}
                                            ,axis=1)
```

```python
df = pd.merge(df,df.energie.apply(pd.Series)
              ,left_index=True, right_index=True             )      
```

```python
df = df.drop(['energielabel','energiemerkmale'],axis=1)
```

```python
numCols = ['anzahl_raeume','grundstuecksflaeche','kaufpreis','wohnflaeche']
df[numCols] = df[numCols].apply(pd.to_numeric,errors='coerce')
```

```python
df['qm_preis'] = df['kaufpreis']/df['wohnflaeche']
```
